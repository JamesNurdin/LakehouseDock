# Deploy Lakehouse

```python
from trino_stack.render import render_collective
from trino_stack.lakehouse import Lakehouse

"""
Deploy lakehouse with given YAML and register to HMS
"""

res = render_collective(
    templates_dir="./trino_stack/manifests",
    values_path="./trino_stack/lakehouses/lakehouse-tpcds.yaml",
)

lh = Lakehouse(res, verbose=False)

lh.deploy()
lh.register_schema("tpcds", "/mnt/iceberg/warehouse", use_dot_db=True) # <schema name in /warehouse>  <location of /warehouse in control container>
```

# Execute Queries/Workloads

```python
from trino_stack.lakehouse import Lakehouse

"""
Issue a single query to the lakehouse
"""
lh = Lakehouse.from_release(instance_name="lakehouse-a", namespace="pgr24james",sync_schemas=True, verbose=False)

try:
    query = generated_query["sql"]
except NameError:
    query = """
    SHOW TABLES FROM iceberg.bigbenchv2_sf1000 LIKE 'web_logs'
    """
    
resullt = lh.issue_query(
    query,
    schema="bigbenchv2_sf1000",
    receive_result=True,
    fetch_ui_doc=False,
    query_plan=False,
    io_plan=False,
)

print(f"{resullt}")
```

```python
from trino_stack.lakehouse import Lakehouse

"""
Run a workload on the lakehouse
"""
NUM_ATTEMPTS = 5 # should be seperate runs, not undersame identifiable run 

lh = Lakehouse.from_release(instance_name="lakehouse-a", namespace="pgr24james",sync_schemas=False, verbose=False)

for i in range(NUM_ATTEMPTS):
    res = lh.run_workload(
        workload_name="gpt_generated_tpcds",
        schema="tpcds",
        attempts=1, # Cache realted
        query=True,
        query_plan = True,
        io_plan = True,
    )
```
